```python
from datetime import datetime

months = {
    "January": "01",
    "February": "02",
    "March": "03",
    "April": "04",
    "May": "05",
    "June": "06",
    "July": "07",
    "August": "08",
    "September": "09",
    "October": "10",
    "November": "11",
    "December": "12"
}

def createISOformat(input: str):
    splitted = input.strip().split()

    if splitted[1] in months:
        month_num = months[splitted[1]]
    
    day_clean, fluff = splitted[2].split("t")

    date_value = datetime(int(splitted[3]), int(month_num), int(day_clean)).isoformat()
    return date_value

createISOformat("Deadline: October 26th, 2025")
```
